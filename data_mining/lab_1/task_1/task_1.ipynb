```python
with open("dune.txt", encoding="UTF-8") as f:
    words = [word for line in f for word in line.split()]
```

```python
with open("stopwords.txt", encoding="UTF-8") as f:
    stopwords = [stopword.rstrip() for stopword in f]
stopwords
```

```python
from string import punctuation
words = [word.lower().translate(str.maketrans('', '', punctuation)) for word in words]
words
```

```python
filtered_words = [w for w in words if not w in stopwords]
filtered_words
```

```python
filtered_words = [w for w in words if not w in stopwords and not len(w)==0]
filtered_words
```

```python
from nltk import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
with open("dune_stemmed.txt", "w") as f:
    for line in stemmed_words:
        f.write(f"{line} ")
stemmed_words
```

```python
from collections import Counter
grouped_pairs = list(Counter(stemmed_words).items())
??Counter
# grouped_pairs
```

```python
grouped_pairs.sort(key=lambda pair: pair[1], reverse=True)
grouped_pairs
```

```python
with open("dune_grouped_pairs.txt", "w") as f:
    for line in grouped_pairs:
        f.write(f"{line}\n")
```
